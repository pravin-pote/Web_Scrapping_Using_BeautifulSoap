# Importing necessary libraries 

```python
# for making http requests to reteive webpage content
import requests

#for parsing and extracting data from HTML
import html5lib

#HTML parser used by BeautifulSoup
from bs4 import BeautifulSoup
```

# Setting the target URL to scrape 

```python
# set the url from where we need to get the data from 
url = "https://www.passiton.com/inspirational-quotes"

# send a GET request to the website
req = requests.get(url)
```

```python
#print response content for debugging(its optional but good for dubugging)
print(req.content)
```

 # Parse the HTML Content

```python
sp = BeautifulSoup(req.content, "html5lib")
```

```python
# Find the main cintainer that holds all quotes
pdiv = sp.find('div', attrs={'id': 'all_items'})
print(pdiv)
```

```python
# to find oall individual quote blocks within the main container
alldivs = pdiv.findAll('div', attrs={'class': 'text-center mb-8'})
```

# Loop for getting Data from the Web quote_&_Author

```python
# now loop through aeach block to extract quote and author 
for d in alldivs:
    allp = d.findAll('p')
    print(allp)
    # Display the author and quote seperately
    print("Author:", allp[1].text,"\n Quote:", allp[0].text)
    print()
```
